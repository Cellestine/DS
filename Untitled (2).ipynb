```python
# Feature Engineering et Prétraitement avancé
# !pip install category_encoders
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from category_encoders import CountEncoder

# 1. Chargement des données nettoyées
train_df = pd.read_csv("/train_cleaned.csv")
test_df = pd.read_csv("/test_cleaned.csv")

binary_cols = [
    'DEROG1', 'DEROG2', 'DEROG3', 'DEROG4', 'DEROG5', 'DEROG6', 'DEROG7', 'DEROG8',
    'DEROG9', 'DEROG10', 'DEROG11', 'DEROG12', 'DEROG13', 'DEROG14', 'DEROG16',
    'KAPITAL34', 'KAPITAL35', 'KAPITAL36', 'KAPITAL37', 'KAPITAL38', 'KAPITAL39',
    'KAPITAL40', 'KAPITAL41', 'KAPITAL42', 'KAPITAL43', 'ADOSS', 'INDEM1', 'INDEM2',
    'RISK6', 'RISK8', 'RISK9', 'RISK10', 'RISK11', 'RISK12', 'RISK13', 'EQUIPEMENT2', 'EQUIPEMENT5'
]

for col in binary_cols:
    if col in train_df.columns:
        # Vérifie si les valeurs sont bien binaires ('O'/'N') avant de remplacer
        unique_vals = train_df[col].dropna().unique()
        if set(unique_vals).issubset({'O', 'N'}):
            train_df[col] = train_df[col].replace({'O': 1, 'N': 0}).astype(float)
        else:
            print(f"⚠️ Attention : '{col}' contient aussi d'autres valeurs : {unique_vals}")

    if col in test_df.columns:
        unique_vals_test = test_df[col].dropna().unique()
        if set(unique_vals_test).issubset({'O', 'N'}):
            test_df[col] = test_df[col].replace({'O': 1, 'N': 0}).astype(float)
        else:
            print(f"⚠️ Attention : '{col}' (test) contient aussi d'autres valeurs : {unique_vals_test}")

print("✅ Conversion des colonnes binaires (O/N → 1/0) terminée.")

surface_cols = [col for col in train_df.columns if "SURFACE" in col.upper()]
capital_cols = [col for col in train_df.columns if "KAPITAL" in col.upper()]
prev_cols = [col for col in train_df.columns if "PREV" in col.upper()]

print("Colonnes surfaces détectées :", surface_cols)
print("Colonnes capitaux détectées :", capital_cols)
print("Colonnes préventions détectées :", prev_cols)

# 3. Création des features

# 1. Surface totale
for col in surface_cols:
    train_df[col] = pd.to_numeric(train_df[col], errors='coerce').fillna(0)
    test_df[col] = pd.to_numeric(test_df[col], errors='coerce').fillna(0)

train_df["surface_totale"] = train_df[surface_cols].sum(axis=1)
test_df["surface_totale"] = test_df[surface_cols].sum(axis=1)

# 2. Capital total
train_df["capital_total"] = train_df[capital_cols].sum(axis=1)
test_df["capital_total"] = test_df[capital_cols].sum(axis=1)

# 3. Grande exploitation (surface > 5000 m²)
train_df["grande_exploitation"] = (train_df["surface_totale"] > 5000).astype(int)
test_df["grande_exploitation"] = (test_df["surface_totale"] > 5000).astype(int)

# 4. Aucune prévention
if prev_cols:
    train_df["aucune_prevention"] = (train_df[prev_cols].sum(axis=1) == 0).astype(int)
    test_df["aucune_prevention"] = (test_df[prev_cols].sum(axis=1) == 0).astype(int)
else:
    print("Aucune colonne 'prev' détectée pour créer 'aucune_prevention'")

# 5. Age du bâtiment
if 'ANNEE_ASSURANCE' in train_df.columns and 'ANCIENNETE' in train_df.columns:
    train_df["age_batiment"] = train_df["ANNEE_ASSURANCE"] - train_df["ANCIENNETE"]
    test_df["age_batiment"] = test_df["ANNEE_ASSURANCE"] - test_df["ANCIENNETE"]

# 6. Surface par bâtiment
train_df["surface_par_batiment"] = train_df["surface_totale"] / (train_df["NBBAT1"] + 1e-5)
test_df["surface_par_batiment"] = test_df["surface_totale"] / (test_df["NBBAT1"] + 1e-5)

# 7. Capital par surface
train_df["capital_par_surface"] = train_df["capital_total"] / (train_df["surface_totale"] + 1e-5)
test_df["capital_par_surface"] = test_df["capital_total"] / (test_df["surface_totale"] + 1e-5)

# 8. Capital moyen par bâtiment
train_df["capital_moyen_par_batiment"] = train_df["capital_total"] / (train_df["NBBAT1"] + 1e-5)
test_df["capital_moyen_par_batiment"] = test_df["capital_total"] / (test_df["NBBAT1"] + 1e-5)

# 9. Taux capital/surface
train_df["taux_capital_surface"] = train_df["capital_total"] / (train_df["surface_totale"] + 1e-5)
test_df["taux_capital_surface"] = test_df["capital_total"] / (test_df["surface_totale"] + 1e-5)

print(" Nouvelles features ajoutées avec succès !")

pd.set_option('display.max_columns', 375)
display(train_df.head())
display(test_df.head())
```

```python
# Remplacer les valeurs dans la colonne AN_EXERC
# 1. Conversion en string (au cas où certains seraient en float ou int)
train_df['AN_EXERC'] = train_df['AN_EXERC'].astype(str)
test_df['AN_EXERC'] = test_df['AN_EXERC'].astype(str)

# 2. Remplacement de 'ANNEEX' par 'X'
# Extraire le chiffre de fin de chaque libellé "ANNEE1", ..., "ANNEE9"
train_df['AN_EXERC'] = train_df['AN_EXERC'].str.extract(r'ANNEE(\d)').astype(int)
test_df['AN_EXERC'] = test_df['AN_EXERC'].str.extract(r'ANNEE(\d)').astype(int)

# 3. Vérification des valeurs uniques restantes (optionnel mais conseillé)
print("Valeurs uniques (train) :", train_df['AN_EXERC'].unique())
print("Valeurs uniques (test)  :", test_df['AN_EXERC'].unique())

pd.set_option('display.max_columns', 375)
display(train_df.head())
display(test_df.head())

# Optionnel : sauvegarde
train_df.to_csv("train_with_features.csv", index=False)
test_df.to_csv("test_with_features.csv", index=False)
```

```python
colonnes_a_verifier = [
    'INDEM2', 'RISK6', 'RISK8', 'RISK9', 'RISK10', 'RISK11', 'RISK12', 'RISK13',
    'EQUIPEMENT2', 'EQUIPEMENT5', 'DEROG12', 'DEROG13', 'DEROG14', 'DEROG16'
]

for col in colonnes_a_verifier:
    if col in train_df.columns and col in test_df.columns:
        valeurs_train = set(train_df[col].dropna().astype(str).unique())
        valeurs_test = set(test_df[col].dropna().astype(str).unique())
        toutes_valeurs = sorted(valeurs_train.union(valeurs_test))
        print(f"\n📌 {col} :")
        print(toutes_valeurs)
```

```python
num_cols = train_df.select_dtypes(include=['number']).columns.tolist()
cat_cols = train_df.select_dtypes(exclude=['number']).columns.tolist()

# Suppression des colonnes ayant plus de 40% de valeurs manquantes ou artificielles (0 pour numériques, "Inconnu" pour catégorielles)
threshold = 0.4
cols_to_drop = []

for col in train_df.columns:
    total = len(train_df)
    if col in num_cols:
        # Pour les numériques, considérer comme "valeurs artificielles" les zéros
        missing_ratio = ((train_df[col] == 0).sum() + train_df[col].isna().sum()) / total
    else:
        # Pour les catégorielles, considérer "Inconnu" comme valeur artificielle
        missing_ratio = ((train_df[col] == "Inconnu").sum() + train_df[col].isna().sum()) / total
    if missing_ratio > threshold:
        cols_to_drop.append(col)

# Supprimer ces colonnes à fort taux de manquants du train et du test
train_df.drop(columns=cols_to_drop, inplace=True)
test_df.drop(columns=[col for col in cols_to_drop if col in test_df.columns], inplace=True)

print(f"Colonnes supprimées (valeurs manquantes > 40%) : {cols_to_drop}")
```

```python
# Identifier les colonnes numériques restantes (après suppression précédente)
num_cols = train_df.select_dtypes(include=['number']).columns.tolist()
cat_cols = train_df.select_dtypes(include=['object']).columns.tolist()

# Calcul de la variance de chaque colonne numérique
variance = train_df[num_cols].var(numeric_only=True)
low_variance_cols = variance[variance < 0.01].index.tolist()

# Suppression des colonnes quasi-constantes (faible variance) du train et du test
train_df.drop(columns=low_variance_cols, inplace=True)
test_df.drop(columns=[col for col in low_variance_cols if col in test_df.columns], inplace=True)

print(f"Colonnes supprimées pour faible variance : {low_variance_cols}")
```

```python
# Calcul de la matrice de corrélation absolue pour les colonnes numériques restantes
corr_matrix = train_df.corr(numeric_only=True).abs()

# Sélection de la partie supérieure de la matrice de corrélation (pour éviter duplications)
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Identifier les colonnes dont la corrélation avec au moins une autre colonne dépasse 0.95
high_corr_cols = [col for col in upper_tri.columns if any(upper_tri[col] > 0.95)]

# Suppression de ces colonnes fortement corrélées du train et du test
train_df.drop(columns=high_corr_cols, inplace=True)
test_df.drop(columns=[col for col in high_corr_cols if col in test_df.columns], inplace=True)

print(f"Colonnes supprimées pour forte corrélation : {high_corr_cols}")

# Vérification finale du nombre de colonnes restantes et de l'absence de NaN
print(f"Nettoyage terminé : {train_df.shape[1]} colonnes conservées après nettoyage")
print("Nombre de valeurs manquantes restantes dans train_df :", train_df.isna().sum().sum())
```

```python
# Importations nécessaires
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
import numpy as np

# Chargement des données préparées (déjà nettoyées)
X_train_full = pd.read_csv('/train_features.csv')
y_train_full = pd.read_csv('/train_output_DzPxaPY.csv')['FREQ']
X_test = pd.read_csv('/test_features.csv')

pd.set_option('display.max_columns', 375)
display(X_train_full.head())
display(y_train_full.head())
display(X_test.head())
```

```python
from sklearn.model_selection import GridSearchCV

# Définir la grille des hyperparamètres à tester
param_grid = {
    'n_estimators': [100,200],
    'max_depth': [3, 5],
    'learning_rate': [0.1,0.05],
    'subsample': [0.6,0.8]
}

# Initialiser le modèle de base
xgb = XGBRegressor(objective="count:poisson", random_state=42)

# GridSearch avec validation croisée
grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    scoring='neg_mean_squared_error',
    cv=3,
    verbose=1,
    n_jobs=-1
)

# Lancer la recherche
grid_search.fit(X_train_full, y_train_full)

# Meilleurs hyperparamètres
best_params = grid_search.best_params_
print("✅ Meilleurs paramètres trouvés :", best_params)

# Remplacer le modèle par le meilleur trouvé
model = XGBRegressor(objective="count:poisson", random_state=42, **best_params)
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error
from xgboost import XGBRegressor
import numpy as np
import pandas as pd

# Split pour évaluation
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.2, random_state=42)

# Modèle avec les meilleurs hyperparamètres
model = XGBRegressor(objective="count:poisson", random_state=42, **best_params)

# Entraînement sur le split
model.fit(X_train, y_train)

# Évaluation sur validation
preds_val = model.predict(X_val)
rmse = np.sqrt(mean_squared_error(y_val, preds_val))
print(f"📉 RMSE validation : {rmse:.4f}")

# Cross-validation sur toutes les données (bonus)
cv_scores = cross_val_score(model, X_train_full, y_train_full, scoring="neg_mean_squared_error", cv=5)
cv_rmse = np.sqrt(-cv_scores)
print(f"📊 Cross-validated RMSE : {cv_rmse.mean():.4f}")

# 🔁 Réentraînement sur toutes les données
model.fit(X_train_full, y_train_full)

# Prédictions sur le test
predictions = model.predict(X_test)

# Soumission
submission = pd.DataFrame({
    "ID": X_test["ID"],
    "FREQ": predictions
})
submission.to_csv("/submission_freq.csv", index=False)
print("✅ Soumission prête : submission_freq.csv")
```
